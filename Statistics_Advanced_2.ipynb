Hypothesis Testing: The CDF is useful for hypothesis testing and calculating critical values.

# Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

Ans3.
The normal distribution / Gaussian distribution is often applied in situations where the data exhibit a bell-shaped, symmetric
pattern around the mean.

Some examples of situations where the normal distribution might be used as a model include:

Height and Weight of Individuals: The distribution of adult heights and weights tends to follow a roughly normal distribution, 
    with most individuals close to the mean height and weight.

Test Scores: In educational testing, the scores of large groups of students on standardized tests often approximate a normal 
    distribution.

Errors in Measurements: When measuring physical quantities with inherent randomness, such as errors in laboratory measurements
    or electronic noise, the distribution of errors is often modeled as normal.

Natural Phenomena: Many natural phenomena, such as the distribution of IQ scores, blood pressure levels, and reaction times,
    approximate a normal distribution.

The normal distribution is characterized by two parameters:

mean (μ)
standard deviation (σ).
Mean (μ):
The mean represents the central value around which the data are centered. It is the location parameter, determining the center
of the distribution. Changing the mean shifts the entire distribution to the left or right without changing its shape.

Standard Deviation (σ):
The standard deviation is a measure of the spread or dispersion of the data points around the mean. It is the scale parameter,
influencing the width of the distribution. A smaller standard deviation results in a narrower distribution, while a larger 
standard deviation results in a broader distribution.

Shape Change with Respect to the Two Paramaters
The shape of the normal distribution remains constant for a given mean, but the spread changes as the standard deviation varies.
A smaller standard deviation makes the distribution more peaked and concentrated around the mean, while a larger standard 
deviation results in a flatter, wider distribution.

When the standard deviation is equal to 1, and the mean is equal to 0, the normal distribution is called the standard normal
distribution. This standardized form allows us to compare and convert any normal distribution to the standard normal 
distribution using z-scores.

# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

The importance of Normal Distribution lies in the following aspects:

Central Limit Theorem: One of the key reasons for the significance of the normal distribution is the Central Limit Theorem 
    (CLT). The CLT states that the sum (or average) of a large number of independent and identically distributed random 
    variables, regardless of their original distribution, will be approximately normally distributed. This property makes
    the normal distribution a fundamental tool for inferential statistics and hypothesis testing.

Well-Understood Properties: The normal distribution is mathematically well-defined and analytically tractable. It allows for
    straightforward calculations and provides easily interpretable measures of central tendency (mean, median, mode) and
    variability (standard deviation).

Approximation for Many Phenomena: Although few real-world data are perfectly normally distributed, many natural processes 
    and social phenomena tend to follow an approximately normal distribution, especially when the sample size is large. The
    normal distribution serves as a useful approximation in such cases.

Statistical Inference: The normal distribution is extensively used in statistical inference, including confidence interval
    estimation, hypothesis testing, and parameter estimation. It forms the basis for several statistical tests, such as t-tests
    and Z-tests.

Examples of Real-Life Normal Distributions:

IQ Scores: IQ (intelligence quotient) scores of a large population tend to follow an approximately normal distribution. The 
    mean IQ is set at 100, and the standard deviation is typically set at 15. This distribution is useful for understanding

Height of Adult Humans: The heights of adult humans in a given population often approximate a normal distribution. The 
    distribution is centered around the average height, with a standard deviation reflecting the variability of heights.

Exam Scores: In educational testing, the distribution of scores on standardized exams, especially when the exam is

# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a single binary outcome in a single trial of 
an experiment. It represents a situation where there are only two possible outcomes: success (usually denoted as "1") with
    probability "p" and failure (usually denoted as "0") with probability "q," where "q = 1 - p."

The Bernoulli distribution is named after the Swiss mathematician Jacob Bernoulli.

The Probability Mass Function (PMF) of the Bernoulli distribution is given by:

P(X = x) = p^x * q^(1-x)

where:

P(X = x) is the probability of observing outcome "x" (either 0 or 1).
"x" takes the value 1 for success and 0 for failure.
"p" is the probability of success.
"q" is the probability of failure.
For Example Consider flipping a fair coin:

The random variable X represents the outcome of the coin flip, where heads is a success (X = 1) and tails is a failure 
(X = 0). Since the coin is fair, the probability of getting heads (success) is 0.5, and the probability of getting tails
(failure) is also 0.5.

The PMF of the Bernoulli distribution for this coin flip is as follows:

P(X = 1) = 0.5 (probability of getting heads)
P(X = 0) = 0.5 (probability of getting tails)

# Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

Ans6.


```python
from scipy.stats import norm

mean = 50
std_dev = 10
observed_value = 60

# calculate the z-score
z_score = (observed_value - mean) / std_dev

# calculate the probability using the cumulative distribution function (cdf)
probability = 1 - norm.cdf(z_score)

print("The probability of selected observation being greater than 60 is:",probability)
```

# Q8: What is the z score? State the importance of the z score.

Ans8.
The z-score, also known as the standard score, is a statistical measure that quantifies the number of standard deviations a 
data point is from the mean of its distribution. It is used to standardize and compare data points across different 
distributions, allowing for a common scale of comparison.

Mathematically, the z-score of a data point "x" in a distribution with mean "μ" and standard deviation "σ" is calculated as:

z = (x - μ) / σ

where:

"x" is the data point whose z-score we want to calculate.
"μ" is the mean of the distribution.
"σ" is the standard deviation of the distribution.
The z-score indicates how many standard deviations a data point is above or below the mean. A positive z-score means the data
point is above the mean, while a negative z-score means it is below the mean.

Importance of the z-score:

Standardization:
The z-score standardizes data, enabling comparisons and evaluations across different distributions. By converting data into
z-scores, we remove the original unit of measurement, allowing for easier interpretation and comparison.

Outlier Detection:
Z-scores are useful for identifying outliers in a dataset. Data points with z-scores significantly larger or smaller than zero
are considered outliers and may require further investigation.

Probability Calculations:
The z-score is used to calculate probabilities and percentiles in the standard normal distribution ( mean = 0, standard
deviation = 1 ) . This simplifies calculations and provides a common reference point.

Hypothesis Testing:
In hypothesis testing, z-scores are used to determine the statistical significance of results. They help assess whether 
observed differences are due to random chance or are statistically meaningful.

Standard Normal Table:
The standard normal distribution (z-distribution) has a well-known cumulative distribution function. Z-scores are used to 
look up probabilities in a standard normal table, making it easier to find critical values for hypothesis testing.

Data Transformation:
Z-scores are often used in data transformation to normalize data and make it suitable for certain statistical analyses, 
such as linear regression.

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans9.
The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that, regardless of the shape of the
original population distribution, the distribution of the sample means will tend to follow a normal distribution as the
sample size increases. In other words, as the sample size becomes larger, the sampling distribution of the sample means
approaches a normal distribution, even if the population distribution is not normal.

The parameters of the sampling distribution of the mean are determined by the parameters of the population:
    
    
    Significance of Central Limit Theorm:

Hypothesis testing : The Central Limit Theorem is used to test hypotheses about the means of populations, where the 
distribution of the population is unknown or non-normal.

Confidence intervals : The Central Limit Theorem is used to construct confidence intervals for the population mean, which 
helps in estimating the true mean of a population based on a sample mean.

Sample size determination : The Central Limit Theorem is used to determine the sample size required for a given level of precision in estimating the population mean.

Quality control : The Central Limit Theorem is used to monitor the quality of production processes by measuring the means of samples and comparing them to the population mean.

# Q10: State the assumptions of the Central Limit Theorem.

Ans10.
Assumptions made of Central Limit Theorem are:

Independence : The samples must be independent of each other. In other words, the value of one sample should not depend on 
    the value of another sample.

Sample size : The sample size should be large enough. Generally, a sample size of at least 30 is considered to be large enough
    for the Central Limit Theorem to hold.

Random sampling : The samples should be selected randomly from the population.

Finite population or replacement : The population should be finite or the sampling should be done with replacement. If the 
    population is infinite and the sampling is done without replacement, then the Central Limit Theorem may not hold.

Finite mean and variance : The population should have a finite mean and variance. If the mean or variance is infinite, then 
    the Central Limit Theorem may not hold.

If all the assumptions are satisified we can then apply Cental Limit Theorem to the data to draw inferences about the 
population parameters and mean.
