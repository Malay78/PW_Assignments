```python
Q3. What is App routing in Flask? Why do we use app routes?
#Ans3.
App routing is the technique used to map the specific URL with associated function intended to perform some task. Modern
web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.

#Use of app routes
If we visit the particular URL mapped to some particular function, th eout of that function is rendered on the browser's
screen, it is used to help users to remember application URLs
```

```python
#Ans5.
The url_for function is used for URL building.
```

```python
#Program for url_for()
from flask import *  
   
app = Flask(__name__)  
  
@app.route('/admin')  
def admin():  
    return 'admin'  
  
@app.route('/librarion')  
def librarion():  
    return 'librarion'  
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))  
    if name == 'librarion':  
        return redirect(url_for('librarion'))  
    if name == 'student':  
        return redirect(url_for('student'))  
if __name__ =='__main__':  
    app.run(debug = True) 
```
